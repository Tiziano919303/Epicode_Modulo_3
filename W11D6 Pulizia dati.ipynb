```python
#Esercizio: lavoro su un df dummy sintetico

#Creo il dataframe
import pandas as pd    #Importo pandas
import numpy as np    #Importo numpy

years = 5

guadagni = pd.DataFrame({
    "Mese": list("GFMAMGLASOND" * years),
    "Anno": np.repeat(list(range(years)), 12),
    "Valore": np.random.randint(800, 5000, 12 * years)
})

guadagni.head()
```

```python
guadagni["Cumsum_totale"] = guadagni["Valore"].cumsum()    #Calcolo la somma cumulativa senza distinzione di anni
guadagni["Cumsum_per_anno"] = guadagni.groupby("Anno")["Valore"].cumsum()    #Calcolo la somma cumulativa per ogni anno
guadagni.head(15)   #Stampo la visualizzazione finale
```

```python
#Esercizio: lavoro sulla tabella DimCustomer del db 'adventureworks' importato da SQL

#Importo la tabella DimCustomer
import pandas as pd
from sqlalchemy import create_engine

engine = create_engine('mysql+pymysql://root:FlAuTi919303@localhost:3306/adventureworks')
df = pd.read_sql("SELECT * FROM DimCustomer", engine)
df.head()
```

```python
#Trasformo nomi in minuscolo e cognomi in maiuscolo
df["Nome_minusc"] = df["FirstName"].str.lower()
df["Cognome_maiusc"] = df["LastName"].str.upper()

#Estraggo nome utente e dominio dalla colonna 'EmailAddress'
df[["Nome_utente", "Dominio"]] = df["EmailAddress"].str.split("@", expand=True)

#Estraggo ogni parte del numero di telefono
df["Parti_numero"] = df["Phone"].str.split(" ")

#Estraggo le email che contengono "21"
df_21 = df[df["EmailAddress"].str.contains("21", na=False)]
```

```python
#Visualizzo risultati e modifiche
df[["Nome_minusc", "Cognome_maiusc", 
    "Nome_utente", "Dominio", 
    "Parti_numero"]].head(11)
```

```python
#Estraggo le mail che contengono 10 o 20
email_20_10 = df[df["EmailAddress"].str.contains("20|10", na=False)]

#Calcolo lunghezze mail e trovo 5 pi√π corte e 5 pi√π lunghe
df["Lungh_mail"] = df["EmailAddress"].str.len()
email_piu_lunghe = df.nlargest(5, "Lungh_mail")[["EmailAddress", "Lungh_mail"]]
email_piu_corte  = df.nsmallest(5, "Lungh_mail")[["EmailAddress", "Lungh_mail"]]

#Sostituisco il dominio "adventure-works.com" con "aw-db.com"
df["Nuovo_dominio"] = df["EmailAddress"].str.replace("adventure-works.com", "aw-db.com")

#Estraggo indirizzo che contengono "Street" nella colonna "AddressLine1"
df_street = df[df["AddressLine1"].str.contains("Street", na=False)]

#Visualizzo risultati e modifiche
print("Email con '20' o '10':")
print(email_20_10[["EmailAddress"]].head())

print("\n5 email pi√π lunghe:")
print(email_piu_lunghe)

print("\n5 email pi√π corte:")
print(email_piu_corte)

print("\nNuovi domini:")
print(df["Nuovo_dominio"].head())

print("\nIndirizzi con 'Street':")
print(df_street[["AddressLine1"]].head())
```
